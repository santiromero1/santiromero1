## 1.1 Sistemas de Ecuaciones Lineales
En este notebook vamos a graficar ecuaciones lineales. Vamos a utilizar numpy para la representación de las ecuaciones y matplotlib para las gráficas. 

### 1.1.1 a) Exactamente una solucion

Creamos vectores para el dominio de las incognitas

```python
x = np.linspace(-6, 6, 10)
x
```

Grafiquemos los siguiente sistema de ecuaciones:

A.- En dos dimensiones

\begin{align}
x - y &= -1 \\
-x + 3y &= 3
\end{align}

```python
y1 = -1 * (-1 - x)
y2 = (3 + x)/ 3.

fig, ax = plt.subplots(1,1, figsize = (10, 5))

# Create the plot
ax.plot(x, y1)
ax.plot(x,y2)
 
# Show the plot
plt.show()
```

Conjunto Solucion:

\begin{align}
x &= 0 \\
y &=  1\\
\end{align}

B.-  En tres dimensiones

\begin{align}
4x - 2y + z &= 11 \\
-2x + 4y - 2z &= -16 \\
x - 2y + 4z &= 17
\end{align}

```python
x, y = np.linspace(-6,6,100), np.linspace(-6,6,100)
X, Y = np.meshgrid(x,y)
```

```python
x = [0,1,2,3,4,5]
```

```python
y = [2,3,4,5,6,7,8]
```

```python
X, Y = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

Z1 = 11 - 4*X + 2*Y
Z2 = (16 - 2*X + 4*Y) / 2
Z3 = (17 - X + 2*Y) / 4

# Plot planes
ax.plot_surface(X,Y,Z1, alpha=0.5, facecolors='r',rstride=100, cstride=100)
ax.plot_surface(X,Y,Z2, alpha=0.5, facecolors='g',rstride=100, cstride=100)
ax.plot_surface(X,Y,Z3, alpha=0.5, facecolors='b', rstride=100, cstride=100)

# Plot solution
ax.plot((1,1),(-8,8),(-9,23), lw=2, c='b')
ax.plot((1,),(-2,),(3,), lw=2, c='k', marker='o')

plt.show()
```

Conjunto Solucion:

\begin{align}
x &= 1 \\
y &= -2 \\
z &= 3 \\
\end{align}

### 1.1.1 c) Infinitas soluciones

B.-  En tres dimensiones

\begin{align}
2x +3y - z &= 0 \\
x + y + 5z &=0 
\end{align}

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

x, y = np.linspace(-6,6,100), np.linspace(-6,6,100)
X, Y = np.meshgrid(x,y)
Z1 = 2*X+3*Y
Z2 = (X+Y)/(-5)

# Plot planes
ax.plot_surface(X,Y,Z1, alpha=0.5, facecolors='r',rstride=100, cstride=100)
ax.plot_surface(X,Y,Z2, alpha=0.5, facecolors='g',rstride=100, cstride=100)

ax.plot((-8,8),(5.5,-5.5),(0.5,-0.5), lw=2, c='b')

plt.show()
```

Conjunto Solucion:

\begin{align}
x &= -16z \\
y &= 11z \\
z &→ \text{libre}\\
\end{align}
